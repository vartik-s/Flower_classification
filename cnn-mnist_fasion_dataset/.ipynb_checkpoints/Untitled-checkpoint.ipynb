# Building CNN using keras
### MNIST fashion data set

```python
#importing libraries
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Input,Convolution2D,MaxPool2D,Flatten,Dense,Dropout
from keras.utils import np_utils
import tensorflow
```

```python
x=pd.read_csv('fashion-mnist.csv')
print(x.head())
x=np.array(x)
X=x[:,1:]
X=X/255.0
y=x[:,0]
```

```python
print(X.shape,y.shape)
```

```python
X_train=X.reshape((-1,28,28,1))
Y_train=np_utils.to_categorical(y)
```

```python
print(X_train.shape,Y_train.shape)
```

```python
Y_train[0]  ###in one hot notation 1 is represented as in the output
```

# Visualize some data

```python
import matplotlib.pyplot as plt
```

```python
for i in range(10):
    plt.imshow(X_train[i].reshape(28,28),cmap='gray')
    plt.show()
```

# CNN model

```python
model=Sequential()
model.add(Convolution2D(32,(3,3),activation='relu',input_shape=(28,28,1)))
model.add(Convolution2D(64,(3,3),activation='relu'))
model.add(Dropout(0.25))
model.add(MaxPool2D(2,2))
model.add(Convolution2D(32,(5,5),activation='relu'))
model.add(Convolution2D(8,(5,5),activation='relu'))
model.add(Flatten())
model.add(Dense(10,activation='softmax'))
model.summary()
```

```python
#compiling model
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
hist=model.fit(X_train,Y_train,epochs=20,shuffle=True,batch_size=256,validation_split=0.2)
```

# training done, time to see how accuracy and loss changes with epochs

```python
plt.plot(hist.history['loss'], label='training loss')
plt.plot(hist.history['val_loss'], label='validation loss')

plt.plot(hist.history['acc'],label='training accuracy')
plt.plot(hist.history['val_acc'],label='validation accuracy')

plt.xlabel('number of epochs')

plt.legend()
plt.show()
```

### As we can see that both the validation and the training loss are decreasing as the number of epochs increases and the the accuracy (both the training as well as the validation accuracy) increases as a function of number of epochs. We finally achieved a validation accuracy of about 87%
